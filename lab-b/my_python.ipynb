### Create t and sin(t) function

```python
sample_start = 0
sample_stop = 100
t = range(sample_start, sample_stop)
```

```python
# Calculate the sine wave
import math

A = 1  # Volts
f = 0.1  # Hz
v_out = [A * math.sin(2 * math.pi * f * time) for time in t]
```

### Plot using px

```python
# Plot the sine wave
import plotly.express as px

fig = px.line(x=t, y=v_out, labels={'x': 'Time [s]', 'y': 'Voltage [V]'})
fig.show()
```

### Check zero-across

```python
A = 1  # Volts
f = 0.1  # Hz

# Find zero crossings
for i, v in enumerate(v_out):
    if v == 0:
        print("Zero crossing at time:", i)
```

```python
# Find zero crossings
for i, v in enumerate(v_out):
    if v < 1e-15 and v > -1e-15:
        print("Zero crossing at time:", i)
```

```python
import cmath
import math

# Potential divider where z2 is a capacitor
f = 16000000  # Hz
w = 2 * math.pi * f  # rad/s

v_in = 5 * cmath.exp(1j * w)

z1 = 1  # kOhm
c = 1e-9  # Farads
z2 = 1 / (1j * w * c)

v_out = (z2 * v_in) / (z1 + z2)  # Volts
vout_mag = abs(v_out)
vout_phase = cmath.phase(v_out)
```
